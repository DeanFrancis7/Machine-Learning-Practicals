Normalization is a scaling technique in which values are shifted and rescaled ranging between 0 and 1. It is called as min-max scalar.

Formula: image.png


Standardization: It is a technique for scaling. In this the values are centered around the mean with a unit standard deviation.

Formula: image.png

If our data is not normally distributed then we use normalization. If our data is normally distributed then we use standardization.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/BostonHousing.csv')
```

```python
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scalar = MinMaxScaler()
normalized_data = scalar.fit_transform(df)
normalized_data
```

```python
new_df = pd.DataFrame(normalized_data, columns=df.columns)
```

```python
new_df
```

```python
new_df.describe()
```

```python
scalar = StandardScaler()
standardized_data = scalar.fit_transform(df)
standardized_data
```

```python
new_df2 = pd.DataFrame(standardized_data, columns=df.columns)
```

```python
new_df2
```

```python
new_df2.describe()
```

Maximum Absolute Scaling : It is used when most of the data is centered distributed.

Formula: image.png


```python
# maximum absolute scaling
from sklearn.preprocessing import MaxAbsScaler
scaler2 = MaxAbsScaler()
max_abs_scale = scaler2.fit_transform(df)
max_abs_scale
```

```python
new_df3 = pd.DataFrame(max_abs_scale, columns=df.columns)
new_df3
```

```python
new_df3.describe()
```

Median and Quantile Scaling Formula : x-median(x)/iqr(x)

```python
# median and quantile scaling
from sklearn.preprocessing import RobustScaler
scaler3 = RobustScaler()
robust_scale = scaler3.fit_transform(df)
robust_scale
```

```python
new_df4 = pd.DataFrame(robust_scale, columns=df.columns)
new_df4
```

```python
new_df4.describe()
```

L2 Regulation is also called as ridge regulation. It provides the over fitting problem by adding penalty to Loss Function. This penalty discourages large coefficients in the model Vector Unit length scaling L2 normalisation (norm)


```python
# L2 regulation
from sklearn.preprocessing import Normalizer
scaler4 = Normalizer()
normal_scale = scaler4.fit_transform(df)
normal_scale
```

```python
new_df5 = pd.DataFrame(normal_scale, columns=df.columns)
new_df5
```

```python
new_df5.describe()
```
